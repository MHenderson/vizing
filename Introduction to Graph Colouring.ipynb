### 1.1 Introduction to Vertex Colouring

The ```vizing``` package has a ```vertex_colouring``` algorithm which can be tailored using visitor object parameters.

````python
import networkx as nx
import vizing
from vizing.colouring import vertex_coloring
````

````python
def standard_colours(G):
    return [G.node[node]['color'] for node in G.node.keys()]
````

````python
G = nx.complete_graph(9)
vertex_coloring(G)
nx.draw_circular(G, node_color = standard_colours(G))
````

````python
G = nx.complete_bipartite_graph(5,6)
vertex_coloring(G)
nx.draw_circular(G, node_color = standard_colours(G))
````

````python
G = nx.petersen_graph()
vertex_coloring(G)
nx.draw_circular(G, node_color = standard_colours(G))
````

````python
G = nx.complete_graph(10)
vertex_coloring(G)
P = nx.drawing.circular_layout(G, scale = 2.5)
D = nx.draw(G, pos = P, node_color = standard_colours(G))
import matplotlib.pyplot as plt
plt.savefig("graph.png")
````

### 1.2 Randomised Greedy Vertex Colouring

The ```RandomOrder``` object can be used to randomise the ordering of vertices coloured by the greedy colouring strategy.

````python
from vizing.colouring import RandomOrder
G = nx.complete_graph(9)
vertex_coloring(G, nodes = RandomOrder)
nx.draw_circular(G, node_color = standard_colours(G))
````

### 1.3 Greedy Vertex Colouring of Random Graphs

````python
G = nx.erdos_renyi_graph(17,0.35)
vertex_coloring(G)
nx.draw_circular(G, node_color = standard_colours(G))
````

3. List Vertex-Colouring
------------------------

````python
G = nx.complete_graph(4)
for node in G.node:
    G.node[node]['list'] = range(4)
G.node
````

````python
from vizing.colouring import InOrder, RandomAvailableColor, vertex_coloring
vertex_coloring(G, nodes = InOrder, choose_color = RandomAvailableColor)
nx.draw_circular(G, node_color = standard_colours(G))
````
